```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
```

# DATA

```python
members=pd.read_csv("C://Users//user//Documents//jupyter assignments//Project//member list banana.csv",sep=",", encoding='cp1252')
members.head()
```

```python
members.isna()
```

```python
members.fillna(value=members, method=None, axis=None, inplace=False, limit=None, downcast=None)
members.head(4629)
```

```python
members = members.dropna(how = 'all', axis=0)
members.head(4625)
```

```python
#make int
members['Fosa Shares']=pd.to_numeric(members['Fosa Shares'])
members['Share Capital']=pd.to_numeric(members['Share Capital'])
members['Normal Shares']=pd.to_numeric(members['Normal Shares'])
members['Member No']=pd.to_numeric(members['Member No'])
members['Dividend']=pd.to_numeric(members['Dividend'])
members['Date of Join']=pd.to_datetime(members['Date of Join'])
members['Date of Birth']=pd.to_datetime(members['Date of Birth'])
members["Join Month"]=pd.DatetimeIndex(members["Date of Join"]).month
members["Join Year"]=pd.DatetimeIndex(members["Date of Join"]).year
members["Join Day"]=pd.DatetimeIndex(members["Date of Join"]).day

members.head()
```

```python
members['Deposits']= members['Normal Shares']+members['Share Capital']+members['Fosa Shares']
members.head()
```

```python
# Get the variable to examine
var = members['Deposits']

# Create a Figure
fig = plt.figure(figsize=(15,20))

# Plot a histogram
members.boxplot(column='Deposits', by='Share Class',)

# Add titles and labels
#plt.title('Deposit Distribution')

# Show the figure
fig.show()
```

```python
mbrshcap=members.groupby(["Join Month", "Join Year"])["Share Capital"].sum().reset_index()
plt.figure(figsize=(20,15))
sns.relplot(x="Join Month", y="Share Capital",palette = 'BrBG_r',
            hue="Join Year", kind="scatter",legend='auto', data=mbrshcap, height=20, aspect=10/15)
plt.title("Monthly Share Capital")
plt.xlabel("Month")
plt.ylabel("Share Capital Deposit")
```

```python
mbrdpst=members.groupby(["Join Month", "Join Year"])["Normal Shares"].sum().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="Join Month", y="Normal Shares", hue="Join Year", data=mbrdpst)
plt.title("Monthly Normal Shares Deposit")
plt.xlabel("Month")
plt.ylabel("Normal Shares Deposit")
```

```python
mbrfosadpst=members.groupby(["Join Month", "Join Year"])["Fosa Shares"].sum().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="Join Month", y="Fosa Shares", hue="Join Year", data=mbrfosadpst)
plt.title("Monthly Fosa Deposit")
plt.xlabel("Month")
plt.ylabel("Fosa Deposit")
```

```python
Mbrs_active=members.groupby(["Join Month", "Join Year"])["Member No"].nunique().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="Join Month", y="Member No", hue="Join Year", data=Mbrs_active)
plt.title("Monthly Active Users")
plt.xlabel("Month")
plt.ylabel("Active Users")
```

```python
members['Deposits']= members['Normal Shares']+members['Share Capital']+members['Fosa Shares']
members.head()
```

```python
members_deposit_avg=members.groupby(["Join Month", "Join Year"])["Deposits"].mean().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="Join Month", y="Deposits", hue="Join Year", data=members_deposit_avg)
plt.title("Monthly Average Deposits ")
plt.xlabel("Month")
plt.ylabel("Deposits")
```
